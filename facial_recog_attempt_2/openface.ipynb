```python
from deepface import DeepFace
```

```python
import os

positive = os.listdir('./data/positive')
negative = os.listdir('./data/negative')

model_name = 'Facenet'
positive
```

```python
import random

for i in range(0, 50):
    pos_image1 = './data/positive/' + positive[random.randint(0, len(positive) - 1)]
    pos_image2 = './data/positive/' + positive[random.randint(0, len(positive) - 1)]
    result = DeepFace.verify(pos_image1, pos_image2, model_name, enforce_detection=False)
    print(f"Similar Test {i}: {result['verified']}")
for i in range(0, 50):
    pos_image = './data/positive/' + positive[random.randint(0, len(positive) - 1)]
    neg_image = './data/negative/' + negative[random.randint(0, len(negative) - 1)]
    result = DeepFace.verify(pos_image, neg_image, model_name, enforce_detection=False)
    print(f"Difference Test {i}: {result['verified']}")
```
